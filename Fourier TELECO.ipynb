```python
from sympy import fourier_series, pi
from sympy import*
import sympy, math, numpy 

```

```python
#São todas formas de chamar as bibliotecas. (RODAR)

#Fazer a Fourier da expressão $exp^{-x}$ quando varia de $-\pi$ a $\pi$.

from os import truncate
from sympy.abc import x
from sympy.plotting import plot
f=exp(-x)
s=fourier_series(f,(x,-pi,pi))
s1 = s.truncate(n=20)
```

```python
s1
```

```python
print(f"A expansão da série de Fourier da função exp(-x) de x indo de -pi a pi é:")
print(s1) #escrita direta

#Plotar a função e a aproimação de Fourier da função exp(-x) de x indo de -pi a pi
p1 = plot(f,(x,-pi,pi), line_color='blue', show=False)
p2 = plot(s1,(x,-pi,pi), line_color='red', show=False)

p1.append(p2[0])
p1.show()

plot(s1)
```
