A série Fourier de uma função periódica de período $2L$ é:

$$\frac {a_0} 2 +\sum _{k=1}^\infty \left( a_k \cos \frac{k\pi x}L
  + b_k \sin \frac{k \pi x}L\right)$$

onde os coeficientes de Fourier são $a_k,\ k=0,1,2,\dots,$ and $b_k,\ k=1,2,3, \dots,$

$$a_k = \frac 1 L \int _{\mbox{--} L}^L f(x) \cos \frac {k\pi x}L\,dx$$

e

$$b_k = \frac 1 L \int _{\mbox{--} L}^L f(x) \sin \frac {k\pi x}L\,dx\,.$$

The nth partial sum of the Fourier series is

$$s_n(x) = \frac {a_0} 2 +\sum _{k=1}^n \left( a_k \cos \frac{k\pi x}L
  + b_k \sin \frac{k\pi x}L\right).$$

```cpp
clear all; close all; clc

syms x k L n

%O próximo comando diz ao MATLAB que k é um número inteiro. Isso permitirá simplificar a chamada (call) de sen e cos para um número inteiro simbólico k.

evalin(symengine, 'assume(k,Type::Integer)');

%Definição das veriáveis e da fórmula da série

a = @(f, x, k, L) int(f*cos(k*pi*x/L)/L,x,-L,L)
b = @(f, x, k, L) int(f*sin(k*pi*x/L)/L,x,-L,L)
fs = @(f,x,n,L) a(f,x,0,L)/2 + symsum(a(f,x,k,L)*cos(k*pi*x/L) + b(f,x,k,L)*sin(k*pi*x/L),k,1,n);

%Input função

f = exp(-x)

%Função que mostra em resultado da função calculada em forma organizada (e não retilínea)

pretty(fs(f,x,20,1))

%

simplify(a(f,x,k,1));
simplify(b(f,x,k,1));

ezplot(fs(f,x,20,1),-1,1)
hold on
ezplot(f,-1,1)
hold off
title('Partial sum with n=20')
```
